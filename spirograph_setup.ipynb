**Epitrochoids and Hypotrochoids for PYNQ Z2**

complex graphs can be generated by rotating a circle around the edge of another circle Hypotrochoids are when the secondary circle is rotated on the inside of the primary circle whereas Epitrochoids are generated by rotating the secondary circle on the outside edge of the primary.

the equations for both are as follows:
Hypotrochoids:
$$
\begin{aligned}&x(\theta )=(R-r)\cos \theta +d\cos \left({R-r \over r}\theta \right)\\&y(\theta )=(R-r)\sin \theta -d\sin \left({R-r \over r}\theta \right)\end{aligned}
$$
where the angle is found with:
$$
2π \times \frac{\operatorname{LCM}(r, R)}{R}
$$

Epitrochoids:
$$
\begin{aligned}&x(\theta )=(R+r)\cos \theta -d\cos \left({R+r \over r}\theta \right)\\&y(\theta )=(R+r)\sin \theta -d\sin \left({R+r \over r}\theta \right)\end{aligned}
$$
where the angle is found with:
$$
2π
$$

for this implementation these angles are split into an array containing equally spaced angles from 0 to the max value for example:
5 points max angle of 4π
would become [0, π, 2π, 3π, 4π]
